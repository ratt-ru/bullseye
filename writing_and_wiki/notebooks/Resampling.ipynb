# UV Coverage

```python
l = NO_ANTENNA
k = NO_ANTENNA
uvw = np.empty([ROW_COUNT,3])
print "Generating coordinates... "
for r in range(0,ROW_COUNT):
        timestamp = r / (NO_BASELINES)
        baseline_index = r % (NO_BASELINES)
        increment_antenna_1_coord = (baseline_index / k)
        
        # calculate antenna 1 and antenna 2 ids based on baseline index using some fancy
        # footwork ;). This indexing scheme will enumerate all unique baselines per
        # timestamp.
        
        l -= (1) * increment_antenna_1_coord
        k += (l) * increment_antenna_1_coord
        antenna_1 = NO_ANTENNA-l
        antenna_2 = NO_ANTENNA + (baseline_index-k)
        new_timestamp = ((baseline_index+1) / NO_BASELINES)
        k -= (NO_BASELINES-NO_ANTENNA) * new_timestamp
        l += (NO_ANTENNA-1) * new_timestamp
        Lx, Ly, Lz = ANTENNA_POSITIONS[antenna_1] - ANTENNA_POSITIONS[antenna_2]
        rotation_in_radians = np.deg2rad(timestamp*INTEGRATION_LENGTH_IN_DEG + REFERENCE_RA)
        sin_ra = np.sin(rotation_in_radians)
        cos_ra = np.cos(rotation_in_radians)
        sin_dec = np.sin(REFERENCE_DEC_RAD)
        cos_dec = np.cos(REFERENCE_DEC_RAD)
        u = -sin_ra*Lx + cos_ra*Ly
        v = -sin_dec*cos_ra*Lx - sin_dec*sin_ra*Ly + cos_dec*Lz
        w = cos_dec*cos_ra*Lx + cos_dec*sin_ra*Ly + sin_dec*Lz
        uvw[r] = [u,v,w]
hrs = int(OBSERVATION_LENGTH_IN_HRS)
mins = int(OBSERVATION_LENGTH_IN_HRS * 60 - hrs*60)
print "Plotting sampling coverage for %dh %dm..." % (hrs,mins)
uvw /= CENTRE_CHANNEL
plt.figure(figsize=(10,10))
plt.title("UV COVERAGE")
plt.plot(uvw[:,0],uvw[:,1],"r+",label="Baselines")
plt.plot(-uvw[:,0],-uvw[:,1],"b+",label="Conjugate Baselines")
plt.xlabel("u ($cycles\cdot rad^{-1}\cdot m^{-1}$)")
plt.ylabel("v ($cycles\cdot rad^{-1}\cdot m^{-1}$)")
plt.legend()
plt.show()
```

# Model Sky setup (simulation / prediction / degridding) step

```python
model_sky = misc.lena() #Talking of beautiful stars... ;)
#model_sky = np.zeros([512,512])
#model_sky[255,255] = 1
model_regular = np.fft.fftshift(np.fft.fft2(np.fft.ifftshift(model_sky))) #numpy's FFT doesn't need normalization
plt.figure(figsize=(10, 10))
plt.title("Model sky")
plt.imshow(model_sky,cmap="gray")
plt.colorbar()
plt.figure(figsize=(10, 10))
plt.title("Amplitudes of real components")
plt.imshow(np.real(model_regular))
plt.colorbar()
plt.figure(figsize=(10, 10))
plt.title("Amplitudes of imaginary components")
plt.imshow(np.imag(model_regular))
plt.colorbar()
plt.show()
```

```python
max_u = np.max(np.abs(uvw[:,0]))
max_v = np.max(np.abs(uvw[:,1]))
print "Maximum u,v: (%f,%f)" % (max_u,max_v)
#N * cell_size_in_rads = 2 * max_uv
#therefore cell_size_in_rads = 2 * max_uv / N
Nx = model_sky.shape[0]
Ny = model_sky.shape[1]
cell_size_u = 2 * max_u / (Nx)
cell_size_v =  2 * max_v / (Ny)
print "Nyquest cell size (radians) in image space (%f,%f)" % (cell_size_u,cell_size_v)
scaled_uv = np.copy(uvw[:,0:2])
scaled_uv[:,0] /= cell_size_u
scaled_uv[:,1] /= cell_size_v
scaled_uv[:,0] += Nx/2
scaled_uv[:,1] += Ny/2
print np.max(scaled_uv[:,0]),np.max(scaled_uv[:,1])
plt.figure(figsize=(10, 10))
plt.title("Sampling of the measurement space (real)")
plt.imshow(np.real(model_regular))
plt.plot(scaled_uv[:,0],scaled_uv[:,1],"w.",label="Baselines")
plt.colorbar()
```

The basic idea behind "degridding" boils down to resampling. It is very similar to upsampling + interpolating a signal - where K taps are inserted between points in all directions and an interpolation method is applied to fill in the blank positions. However, here we're adding one tap (the continuous uv coord) somewhere between grid taps. Since the images are already critically sampled according to the size of the telescope array there will be no higher frequency spikes between grid points. The goal in degridding is not to add high frequency terms, so just as with upsampling we need to "convolve through" with a simple low pass filter.

```python
#filter options
SUPPORT = 3 #half support
OVERSAMPLE = 5
FULL_SUPPORT = SUPPORT * 2 + 1
PADDED_FULL_SUPPORT = FULL_SUPPORT + 2 # one tap on either side
FILTER_SIZE = (PADDED_FULL_SUPPORT) + (PADDED_FULL_SUPPORT - 1) * (OVERSAMPLE - 1) #filter array dimension size
FILTER_CENTRE = (FILTER_SIZE) // 2
x = np.arange(-FILTER_CENTRE,FILTER_CENTRE + 1)
AA = np.sinc(x/float(OVERSAMPLE)) #we should maybe look at more intricate filter designing than this...
plt.figure()
plt.plot(x,AA)
plt.plot([0,0],[np.max(AA),np.min(AA)])
plt.show()
```

```python
#Speed restrictions ahead. Metrorail apologizes for the inconvenience
vis = np.zeros([ROW_COUNT],dtype=np.complex)
for r in range(0,ROW_COUNT):
    disc_u = int(round(scaled_uv[r,0]))
    disc_v = int(round(scaled_uv[r,1]))
    frac_u_offset = (1 - scaled_uv[r,0] + disc_u) * OVERSAMPLE
    frac_v_offset = (1 - scaled_uv[r,1] + disc_v) * OVERSAMPLE
    if (disc_v + FULL_SUPPORT  >= Ny or disc_u + FULL_SUPPORT >= Nx or
        disc_v < 0 or disc_u < 0): 
        continue
    interpolated_value = 0.0 + 0.0j
    norm_value = 0
    for conv_v in range(0,FULL_SUPPORT):
         v_tap = AA[conv_v * OVERSAMPLE + frac_v_offset]  
         for conv_u in range(0,FULL_SUPPORT):
            u_tap = AA[conv_u * OVERSAMPLE + frac_u_offset]
            conv_weight = v_tap * u_tap
            interpolated_value += model_regular[disc_u - SUPPORT + conv_u, disc_v - SUPPORT + conv_v] * conv_weight
            norm_value += conv_weight
    vis[r] = interpolated_value / norm_value
```

# Image synthesis / gridding step

```python
#enable this if you want to get compute the psf - and overwrite whatever the degridder gave us:
#(very nice way to compare outputs between true and predicted visibilities, btw)
#vis[:] = 1.0 + 0.0j
measurement_regular = np.zeros([Nx,Ny],dtype=np.complex)
norm_value = 0
for r in range(0,ROW_COUNT):
    disc_u = int(round(scaled_uv[r,0]))
    disc_v = int(round(scaled_uv[r,1]))
    frac_u_offset = (1 - scaled_uv[r,0] + disc_u) * OVERSAMPLE
    frac_v_offset = (1 - scaled_uv[r,1] + disc_v) * OVERSAMPLE
    if (disc_v + FULL_SUPPORT  >= Ny or disc_u + FULL_SUPPORT >= Nx or
        disc_v < 0 or disc_u < 0): 
        continue
    for conv_v in range(0,FULL_SUPPORT):
         v_tap = AA[conv_v * OVERSAMPLE + frac_v_offset]  
         for conv_u in range(0,FULL_SUPPORT):
            u_tap = AA[conv_u * OVERSAMPLE + frac_u_offset]
            conv_weight = v_tap * u_tap
            measurement_regular[disc_u - SUPPORT + conv_u, disc_v - SUPPORT + conv_v] += vis[r] * conv_weight
            norm_value += conv_weight
measurement_regular /= norm_value
dirty_sky = np.fft.fftshift(np.fft.ifft2(np.fft.ifftshift(measurement_regular))) * (model_sky.shape[0]*model_sky.shape[1]) #undo numpy's IFFT normalization... ffs...
plt.figure(figsize=(10, 10))
plt.title("Dirty map")
plt.imshow(np.real(dirty_sky),cmap="gray")
plt.colorbar()
plt.show()
```
